##Exercise 1:


```python
product1 = int(input("Input 1° product: "))
product2 = int(input("Input 2° product: "))
product3 = int(input("Input 3° product: "))
product4 = int(input("Input 4° product: "))
product5 = int(input("Input 5° product: "))
sum = (product1 + product2 + product3 + product4 + product5) 
iva = sum * 0.19
total = sum + iva
print("Valor bruto: ", sum, "\nIva: ", iva, "\nTotal: ", total)
```

###Python version


```python
product1, product2, product3, product4, product5 = input("input 1°, 2°, 3°, 4° and 5° product with , separate: ").split(",")
sum = (product1 + product2 + product3 + product4 + product5)
print("Valor bruto:", sum , "\nIva:", (sum * 0.19), "\nTOTAL: ", (sum+(sum*0.19)))
```

##Exercise 2

```python
number1 = int(input('Enter the 1 number '))
number2 = int(input('Enter the 2 number '))
number3 = int(input('Enter the 3 number '))

sum = number1 + number2 + number3 
average = sum / 3

if(number1 > number2 and number1 > number3):
  print('The largest number is: ', number1)
elif(number2 > number1 and number2 > number3):
  print('The largest number is: ', number2)
else:
  print('The largest number is:', number3)

if(number1 < number2 and number1 < number3):
  print('The minimun number is: ', number1)
elif(number2 < number1 and number2 < number3):
  print('The minimun number is: ', number2)
else:
  print('The minimun number is:', number3)

print('Average: ', average)
```
